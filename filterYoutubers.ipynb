```python
import pandas as pd

data = pd.read_csv('youtubeChannelData.csv')
data.head()
```

```python
category_counts = data['Category'].value_counts()
category_counts
```

```python
unclassified_channels = data[data['Category'] == '미분류']
unclassified_channels
```

```python
data = data.drop_duplicates(subset=['Channel Name', 'Channel ID'], keep='first')
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data.reset_index(drop=True, inplace=True)
data
```

```python
data.to_csv('youtubeChannelData.csv', encoding="utf-8-sig")
```
