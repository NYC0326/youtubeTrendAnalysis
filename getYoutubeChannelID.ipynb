```python
import requests, re
import pandas as pd
from bs4 import BeautifulSoup

# 헤더 설정
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
```

```python
df = pd.DataFrame(columns=['Channel Name', 'Channel ID', 'Category'])

# 구독자가 10000명 이상인 한국 유튜버 한정으로 데이터 수집
for i in range(1, 1042):
    start_url = f"https://socialerus.com/Ranking/Index?PageNo={i}&ChCategory=&ChTheme=&ChSubscMin=&ChSubscMax=&ChViewMin=&ChViewMax=&ChVideoMin=&ChVideoMax=&Sort=subsc"
    req = requests.get(start_url, headers=header)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    ranking_infos = soup.find_all("div", {"class": "ranking_info"})

    # 유튜브 채널 ID, 채널명, 카테고리를 크롤
    for info in ranking_infos:
        onclick_attr = info.get('onclick')
        channel_id = re.search(r"id=([^']+)", onclick_attr).group(1) if onclick_attr else None
        category = info.find('span', class_='category').text.strip() if info.find('span', class_='category') else None
        name = info.find('span', class_='title').text.strip() if info.find('span', class_='title') else None

        new_data = pd.DataFrame([[name, channel_id, category]], columns=['Channel Name', 'Channel ID', 'Category'])
        df = pd.concat([df, new_data], ignore_index=True)
df
```

```python
# 중복된 데이터가 있어 삭제
df = df.drop_duplicates(subset=['Channel Name', 'Channel ID'], keep='first')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.reset_index(drop=True, inplace=True)
df
df.to_csv('youtubeChannelData.csv', encoding="utf-8-sig")
```
